```python
topic_rec=df_data_1.drop(['Unnamed: 0'], axis=1)
topic_rec.head()
```

```python
X1=topic_rec[['Text']]
X1=X1['Text'].tolist()
```

#### Bag of words módszerrel a cikkek átfordítása 0/1 számokra

```python
import keras
from keras.preprocessing.text import Tokenizer
tokenizer=Tokenizer(num_words=3000,lower=True)
tokenizer.fit_on_texts(X1)
```

```python
X=tokenizer.texts_to_matrix(X1, mode='binary')
X.shape
```

#### A szöveges célváltozók cseréje számokra

```python
topic_rec=topic_rec.replace('techtud', 0)
topic_rec=topic_rec.replace('kultur', 1)
topic_rec=topic_rec.replace('sport', 2)
topic_rec.head()
```

```python
topic_rec.Topic.unique()
```

```python
Y=topic_rec[['Topic']].values
Y.shape
```

#### Train és Test felbontás

```python
from sklearn.model_selection import train_test_split
X_trainset, X_testset, Y_trainset, Y_testset = train_test_split(X, Y, test_size=0.3, random_state=3)
X_trainset.shape
```

```python
Y_trainset.shape
```

#### Modellezés 

```python
from keras.utils import to_categorical


model = keras.Sequential([
    keras.layers.Dense(32, input_shape=(3000,)  ,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(3, activation='softmax')
    
])
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_trainset, to_categorical(Y_trainset), epochs=20)
```

#### Modell kiértékelése

```python
model.evaluate(
  X_testset,
  to_categorical(Y_testset)
)
```

#### Paraméterek amit tune-oltam, hogy minél jobb eredmény legyen teszt mintán:
- n -> különböző  leggyakoribb szavak száma ami jön a bag of words-ből
- neurális háló struktúrája -> kell-e penalizaton/hány neuron legyen
- epoch szám

#### Többszöri próbálkozás eredménye összegezve:
- 3000-nél sokkal alacsonyabb n jelentősen ront a modell teljesítményén, de nagyobb n nem tud javítani
- penalization javít a modell teljesítményén ez Dropout-tal lett megoldva
- 32 neuron ideális egy layerben -> 16-al még gyengébb a teljesítmény, 64-nél már nem javul
- epoch szám majdnem mindegy ha elég magas, mert az early stopping-ot mint overfit megakadályozása most nem használtam
- a legjobb teszt minta eredmények mind 92-94% körül voltak
- 99,5% a train mintán a teljesítmény, amitől gyanús, hogy overfit a modell, de kevésbé overfit modellek (96, 97% pontosságúak train mintán) sem értek el jobb teljesítményt teszt mintán
